```python
import pandas as pd
import plotly.graph_objects as go

# Read your CSV file
df = pd.read_csv('world_data_clean.csv')

# Convert GDP columns to numeric (they're currently strings)
df['gdp_pc_2013'] = pd.to_numeric(df['gdp_pc_2013'], errors='coerce')
df['gdp_pc_2023'] = pd.to_numeric(df['gdp_pc_2023'], errors='coerce')
df['gdp_2013'] = pd.to_numeric(df['gdp_2013'], errors='coerce')
df['gdp_2023'] = pd.to_numeric(df['gdp_2023'], errors='coerce')

# Create the figure
fig = go.Figure()

# Define metrics to loop through
# Format: (column_name, display_name, color_scale, title, colorbar_title)
metrics = [
    ('gdp_pc_2013', 'GDP per capita 2013', 'Blues', 'GDP per Capita by Country (2013)', 'GDP per capita<br>(USD)'),
    ('gdp_pc_2023', 'GDP per capita 2023', 'Blues', 'GDP per Capita by Country (2023)', 'GDP per capita<br>(USD)'),
    ('gdp_2013', 'GDP 2013', 'Greens', 'Total GDP by Country (2013)', 'GDP<br>(USD)'),
    ('gdp_2023', 'GDP 2023', 'Greens', 'Total GDP by Country (2023)', 'GDP<br>(USD)')
]

# Loop through metrics and create a trace for each
for i, (col, name, color, title, colorbar_title) in enumerate(metrics):
    fig.add_trace(go.Choropleth(
        locations=df['Country Code'],
        z=df[col],
        text=df['Country Name'],
        colorscale=color,
        autocolorscale=False,
        reversescale=False,
        marker_line_color='darkgray',
        marker_line_width=0.5,
        colorbar_title=colorbar_title,
        visible=(i == 0),  # Only first trace visible by default
        name=name
    ))

# Create dropdown buttons
buttons = []
for i, (col, name, color, title, colorbar_title) in enumerate(metrics):
    # Visibility array: True for current metric, False for others
    visibility = [False] * len(metrics)  # Start with all False
    visibility[i] = True  # Set current one to True
    
    buttons.append(
        dict(
            label=name,
            method="update",
            args=[{"visible": visibility},
                  {"title": title}]
        )
    )

# Add dropdown menu to layout
fig.update_layout(
    updatemenus=[
        dict(
            active=0,
            buttons=buttons,
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1.15,
            yanchor="top"
        ),
    ]
)

# Update the layout
fig.update_layout(
    title_text='GDP per Capita by Country (2013)',
    title_x=0.5,
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type='equirectangular'
    ),
    height=600,
    width=1000
)

# Display in notebook (optional)
fig.show()

#  Save as HTML for your website
fig.write_html("/Users/arumehta/Desktop/DS 4200/map_viz.html")
```
